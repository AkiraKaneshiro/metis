### Data exploration with the top 100 grossing movies from 2013 

```python
import pandas as pd
import numpy as np
from datetime import datetime
import dateutil.parser

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
movie_df = pd.read_csv('2013_movies.csv', parse_dates={'Date':['ReleaseDate']})

movie_df.head()
```

#### Domestic total gross over time

Are there seasonality trends in domestic total gross?

```python
movie_df['Month'] = map(lambda x: x.strftime('%m'), movie_df['Date'])
```

```python
sb.factorplot('Month', 'DomesticTotalGross', data=movie_df, palette='PuBu_d', kind='bar');
```

```python
# Line graph
plt.figure(figsize=(10,5))
movie_df = movie_df.sort(['Date'])
plt.plot(movie_df['Date'], movie_df['DomesticTotalGross'])
plt.title('2013 Top Grossing Movies: Domestic Gross by Release Date')
plt.xticks(rotation=45)
plt.xlabel('Release Date')
plt.ylabel('Domestic Total Gross ($)')
```

#### Runtime vs domestic total gross

Does a movie's duration affect domestic total gross?

```python
sb.lmplot('Runtime', 'DomesticTotalGross', data=movie_df, ci=None)
```

#### What is the average runtime and domestic total gross by MPAA rating?

```python
movie_df['DomesticTotalGross'].describe()
```

```python
# pd.options.display.float_format='{:.1f}'.format

movie_df.groupby('Rating')['Runtime', 'DomesticTotalGross'].mean()
```

#### Plots of release date and domestic total gross by MPAA ratings

```python
date_gross_plot = sb.FacetGrid(movie_df, col='Rating')
date_gross_plot.map(plt.plot, 'Date', 'DomesticTotalGross').set_xticklabels(rotation=45)
date_gross_plot.add_legend();
```

```python
date_gross_plot = sb.FacetGrid(movie_df, hue='Rating', size=10)
date_gross_plot.map(plt.plot, 'Date', 'DomesticTotalGross')
date_gross_plot.add_legend(title="Ratings");
```
