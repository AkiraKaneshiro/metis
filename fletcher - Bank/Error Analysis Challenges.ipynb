For the house representatives data set, calculate the accuracy, precision, recall and f1 scores of each classifier you built (on the test set).

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

from sklearn import cross_validation
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support
from sklearn.cross_validation import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
party_df = pd.read_csv('house-votes-84.data.txt', names=['party', 'infants', 'water', 'adoption', 'physician',\
                                                         'elsalvador', 'religious', 'antisatellite', 'nicaraguan',\
                                                         'missile', 'immigration', 'synfuels', 'education', 'superfund', \
                                                         'crime', 'dutyfree', 'export'], header=None)

#    1. Class Name: 2 (democrat, republican)
#    2. handicapped-infants: 2 (y,n)
#    3. water-project-cost-sharing: 2 (y,n)
#    4. adoption-of-the-budget-resolution: 2 (y,n)
#    5. physician-fee-freeze: 2 (y,n)
#    6. el-salvador-aid: 2 (y,n)
#    7. religious-groups-in-schools: 2 (y,n)
#    8. anti-satellite-test-ban: 2 (y,n)
#    9. aid-to-nicaraguan-contras: 2 (y,n)
#   10. mx-missile: 2 (y,n)
#   11. immigration: 2 (y,n)
#   12. synfuels-corporation-cutback: 2 (y,n)
#   13. education-spending: 2 (y,n)
#   14. superfund-right-to-sue: 2 (y,n)
#   15. crime: 2 (y,n)
#   16. duty-free-exports: 2 (y,n)
#   17. export-administration-act-south-africa: 2 (y,n)
```

```python
# change votes to binary values and keep missing votes as NaN

party_df.replace(to_replace = ("y", "n", "?"), value = (1, 0, np.nan), inplace = True)
```

```python
from numpy import nan

party_df.mean().fillna(0)
party = party_df.fillna(party_df.mean().fillna(0))
```

### Other Classifiers

```python
from sklearn import naive_bayes
from sklearn.metrics import accuracy_score, classification_report

X = party[cols]
y = party['party'] 

# split vote data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y)

classifiers = {'lr':LogisticRegression(), 
               'rf':RandomForestClassifier(n_estimators = 200), 
               'dt':DecisionTreeClassifier(),
               'bernNB':naive_bayes.BernoulliNB(),
               'gaussNB':naive_bayes.GaussianNB(),
               'multiNB':naive_bayes.MultinomialNB(),
               'SVC':SVC()
               }

for key, classifier in classifiers.iteritems():
    model = classifier
    model.fit(X_train, y_train)

    print key
    print 'Accuracy: %.3f' % accuracy_score(y_test, model.predict(X_test))
    print classification_report(y_test, model.predict(X_test))
```

```python
# using cross-validation

X = party[cols]
y = party['party'] 

print 'accuracy score'
print 'Logistic', np.mean(cross_val_score(LogisticRegression(), X, y, cv=5))
print 'GaussianNB', np.mean(cross_val_score(GaussianNB(), X, y, cv=5))
print 'BernoulliNB', np.mean(cross_val_score(naive_bayes.BernoulliNB(), X, y, cv=5))
print 'MultinomialNB', np.mean(cross_val_score(naive_bayes.MultinomialNB(), X, y, cv=5))
print 'SVC', np.mean(cross_val_score(SVC(kernel='linear', probability=True), X, y, cv=5))
print 'DecisionTreeClassifier', np.mean(cross_val_score(DecisionTreeClassifier(), X, y, cv=5))
print 'RandomForestClassifier', np.mean(cross_val_score(RandomForestClassifier(), X, y, cv=5))
```

### Challenge 2

For each, draw the ROC curve and calculate the AUC.

```python
import pylab as pl
from sklearn import svm
from sklearn.metrics import roc_curve, auc
```

```python
# single ROC curve on SVC classifier, 
# for different classifiers, change classifier variable
# for example, classifier = LogisticRegression()

democrat = party['party'].map(lambda x: 1 if x == 'democrat' else 0)

X = party[cols]
# y = party['party']
y = democrat 

X_train, X_test, y_train, y_test = train_test_split(X, y)

classifier = svm.SVC(kernel='linear', probability=True)
probas_ = classifier.fit(X_train, y_train).predict_proba(X_test)

# Compute ROC curve and area the curve
fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
roc_auc = auc(fpr, tpr)

# Plot ROC curve

fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)

plt.plot(fpr, tpr, color='lightblue', lw=2, label='ROC curve')
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='dotted', label='random guessing')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC)')
plt.legend(loc="lower right")
ax.annotate('AUC = %0.2f' %roc_auc, xy=(0.35, 0.6))
plt.show()
```

```python
# ROC curve for multiple classifiers

classifiers = {'lr':LogisticRegression(), 
               'rf':RandomForestClassifier(), 
               'dt':DecisionTreeClassifier(),
               'knn':KNeighborsClassifier(),
               'svc':SVC(kernel='linear', probability=True),
               'bernNB':naive_bayes.BernoulliNB(),
               'gaussNB':naive_bayes.GaussianNB(),
               'multiNB':naive_bayes.MultinomialNB()
               }

for name, classifier in classifiers.iteritems():
    
    cv = cross_validation.StratifiedKFold(y_test, n_folds = 5)

    for train, test in cv:
        probas_ = classifier.fit(X_train, y_train).predict_proba(X_test)
        fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])

    roc_auc = auc(fpr, tpr)
    pl.plot(fpr, tpr, lw=1, label='ROC %s (area = %0.2f)' % (name, roc_auc))
    
pl.plot([0, 1], [0, 1], '--', color=(0.6, 0.6, 0.6), label='Luck',)
pl.xlim([-0.05, 1.05])
pl.ylim([-0.05, 1.05])
pl.xlabel('False Positive Rate')
pl.ylabel('True Positive Rate')
pl.title('ROC Curve')
pl.legend(loc="lower right")
pl.show()
```
