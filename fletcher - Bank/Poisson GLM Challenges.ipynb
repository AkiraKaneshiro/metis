### Challenge 1
Model the damage incident counts with a Poisson Regression.

```python
# from pandas.io.stata import StataReader
# reader = StataReader('ships.dat')
# dataframe = reader.data()

import pandas as pd
df = pd.read_stata('ships.dta')
```

```python
df.describe()
```

```python
# dummifying ship type, construction year, and operation year
ship_type = pd.get_dummies(df['type'])
df = df.join(ship_type)

construction = pd.get_dummies(df['construction'])
df = df.join(construction)

operation = pd.get_dummies(df['operation'], prefix='operation')
df = df.join(operation)

# adding constant
df['constant'] = 1.0
```

```python
df 
```

```python
df.columns
```

```python
import statsmodels.api as sm
from statsmodels.discrete.discrete_model import Poisson

bin_Y = df['damage']
X = df[['months', 'A', 'B', 'C', 'D', '1960-64', '1965-70', \
        '1970-74', 'operation_1960-74']]
# left out E, '1975-79', 'operation_1975-79'

sm_X = sm.add_constant(X) # required with statsmodel

glm_logit = sm.GLM(bin_Y, sm_X, family=sm.families.Poisson())
# sm.families.links.log - default
glm_logit_results = glm_logit.fit()

print glm_logit_results.summary()

# Interpretation:
# For each coefficient, take e^Beta 
# When months increases by a factor of one, damage 0.50
# Ship type A has 1.16 times more damage compared to ship type E
# Ship type B has 2.26 times more damage compared to ship type E
# Ship type C has 0.35 times more damage compared to ship type E
# Ship type D has 0.51 times more damage compared to ship type E
```
