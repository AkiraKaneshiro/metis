```python
# working code with irmak

url = "http://en.wikipedia.org/wiki/List_of_children's_books_made_into_feature_films"

# Retrieve text from Wikipedia tables 
def get_individual_movie_book():
    book_movie_list = []
    book_movie_dict = {}
    response = requests.get(url)
    soup = BeautifulSoup(response.text.encode('utf-8').decode('ascii', 'ignore'))
    for i in range(23): # builds list of lists
        for link in soup('table')[i].find_all('td'):
            book_split = book_movie_list.append(link.text.split('\n'))
        
    indiv_book_movie_list = []        
    for i, entry in enumerate(book_movie_list):
        if i % 2 == 0: # evens = author, book info
            book_info = entry
        else: # odds = corresponding film
            for indiv_movie in entry:
                indiv_movie_dict = {} # new dict for each key/value pair to be book and movie info
                indiv_movie_dict['book_info'] = book_info
                indiv_movie_dict['movie_info'] = indiv_movie
                indiv_book_movie_list.append(indiv_movie_dict)
                
    return indiv_book_movie_list

        
# pprint(get_individual_movie_book())


# see if it is necessary to clean up book info 

# to match movie with book info
# leftjoin.lower.strip() == rightjoin.lower.strip()
```
