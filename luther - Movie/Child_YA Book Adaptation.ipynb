```python


def get_movie_value(soup, field_name):
    obj = soup.find(text=re.compile(field_name))
    
    if not obj:     # if you can't find anything (falsey), return None
        return None
    
    next_sibling = obj.findNextSibling()
    if next_sibling: # if truthy
        return next_sibling.text
    else:
        return None

def get_gross(soup, field_name):
    gross=None
    rev = soup.find_all(class_='mp_box_content')
    for td in rev[0].find_all('td'):
        obj = re.search(field_name, td.text)
        if obj:
            gross = money_to_int(td.nextSibling.nextSibling.text)
    return gross

# def get_domestic_gross(soup):
#     domestic_gross=None
#     domestic_rev = soup.find_all(class_='mp_box_content')
#     for td in domestic_rev[0].find_all('td'):
#         domestic = re.search(r'Domestic', td.text)
#         if domestic:
#             domestic_gross = money_to_int(td.nextSibling.nextSibling.text)
#     return domestic_gross

# working code for pulling domestic total gross from top table on box office mojo
#     raw_domestic_total_gross = get_movie_value(sub_soup, 'Domestic Total')
#     domestic_total_gross = money_to_int(raw_domestic_total_gross)


# parse date
def to_date(datestring):
    try:
        date = dateutil.parser.parse(datestring)
        return date
    except:
        return None

def money_to_int(moneystring):
    try:
        moneystring = moneystring.replace('$','').replace(',','')            
        return int(moneystring)
    except:
        return None

def text_to_num(moneystring):
    word_number_dict = {
                    'million': 6,
                    'billion': 9
                    }
    word_number = moneystring[-7:]
    if word_number in word_number_dict:
        number = moneystring[:-8].replace('$','')
        return int(float(number) * 10 ** word_number_dict[word_number])
    else:
        return moneystring
    
def get_widest_release(soup):
    widest_release = None
    try:
        sub_widest_release = soup.find_all(class_='mp_box_content')    
        for td in sub_widest_release[1].find_all('td'):
            widest = re.search(r'Widest\xa0Release:', td.text) 
            if widest:
                widest_release = money_to_int(td.nextSibling.nextSibling.text.split()[0])
                return widest_release
    except:
        return widest_release   

# measured in days        
def get_in_release(soup):
    in_release = None
    try:
        sub_in_release = soup.find_all(class_='mp_box_content')
        for td in sub_in_release[1].find_all('td'):
            in_rel = re.search(r'In Release:', td.text)      
            if in_rel:
                in_release = money_to_int(td.nextSibling.nextSibling.text.split()[0])
                return in_release
    except: 
        return in_release
        
def runtime_to_minutes(runtimestring):
    runtime = runtimestring.split() # defaults to white space
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None


def get_director(soup):    
    urls_test = soup.find_all('a')
    for link in urls_test:
        link_string = link.get('href')
#         sub_director = re.search(r'people/chart/\?view=Director', link_string)
        sub_director = re.search(r'people/chart/\?view=Director', link_string)
        if sub_director:
            director = link.text
            return director
#         else:
#             return None

        
# directors without hyperlinks - NEEDS WORK - if director = None       
# def get_older_director(soup):
# #     if get_director(soup) == 'None':
#         ele=soup('font', size="2")
#         return ele[2].contents[0]

# def title(soup):
#     title_string = soup.find('title').text
#     title = title_string.split('(')[0].strip() 
#     return title

```

```python
# opening each link and building features

root_url = "http://boxofficemojo.com"
movie_urls = individual_movie_urls() + additional_indiv_movie_urls 

# def get_individual_movie_data(urls):

movie_data = []

for urls in movie_urls:
    movie_dict = {}
    sub_response = requests.get(root_url + urls)
    sub_soup = BeautifulSoup(sub_response.text)
    
    # title
    title_string = sub_soup.find('title').text
    title = title_string.split('(')[0].strip()  
    movie_dict['movie_title'] = title    
    
    # distributor
    distributor = get_movie_value(sub_soup, 'Distributor')
    movie_dict['distributor'] = distributor  
    
    # genre
    genre = get_movie_value(sub_soup, 'Genre:')
    movie_dict['genre'] = genre     
    
    # domestic total
    domestic_gross = get_gross(sub_soup, 'Domestic')
    movie_dict['domestic_total_gross'] = domestic_gross
    
    # domestic percentage
    
    # foreign total
    foreign_gross = get_gross(sub_soup, 'Foreign')
    movie_dict['foreign_total_gross'] = foreign_gross

    # foreign percentage
    
    # worldwide
    worldwide_gross = get_gross(sub_soup, 'Worldwide')
    movie_dict['worldwide_total_gross'] = worldwide_gross

    # runtime
    raw_runtime = get_movie_value(sub_soup, 'Runtime')
    runtime = runtime_to_minutes(raw_runtime)
    movie_dict['runtime'] = runtime
    
    # rating 
    rating = get_movie_value(sub_soup, 'MPAA Rating') 
    movie_dict['rating'] = rating
    
    # release date
    raw_release_date = get_movie_value(sub_soup, 'Release Date')
    release_date = to_date(raw_release_date)
    movie_dict['release_date'] = release_date
    
    # production budget
    raw_budget = get_movie_value(sub_soup, 'Production Budget')
    budget = text_to_num(raw_budget)
    movie_dict['budget'] = budget 
    
    # widest release number
    widest_release = get_widest_release(sub_soup)
    movie_dict['widest_release'] = widest_release
    
    # in release, days 
    in_release = get_in_release(sub_soup)
    movie_dict['in_release'] = in_release
    
    # director
    director = get_director(sub_soup)
    movie_dict['director'] = director

    movie_data.append(movie_dict)
        
```
