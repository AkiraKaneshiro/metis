```python
import pymongo
```

```python
c = pymongo.MongoClient()
db = c['chocolatedb']
tweets = db['tweets']
```

```python
tweets.count()
```

```python
import re

# save tweet text
all_text = ""
tweet_list = []
for tweet in tweets.find():
    text = tweet['text'].lower()
    text = re.sub('@\S+', ' ', text)  # Remove mentions
    text = re.sub('http\S+', ' ', text)  # Remove urls
    tweet_list.append(text) # Retain words, TfidfVectorizer requires a list
    all_text += text + "\n" # TextBlob requires str
```

```python
# vectorize tweets for KMeans
import numpy as np
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer 

stop = stopwords.words('english')
stop += ["chocolate", "rt"]
         
vectorizer = TfidfVectorizer(stop_words=stop)
# print tweet_list[0]

tweet_vectors = vectorizer.fit_transform(tweet_list) 
print tweet_vectors
```

```python
# examine the features

from pprint import pprint
vocab_index = vectorizer.vocabulary_   # vocab_index is a single dict 

# append all features and indices into a dict
index_vocab = {index: feature for feature, index in vocab_index.items()}
# print 'index, word:', index_vocab.items()[:5] # examine the first 5 pairs


# translate sparse to coo matrix

from scipy.sparse import coo_matrix
from pprint import pprint


top_tweet_list = []

for i in range(len(tweet_list)):
    individual_tweet = tweet_vectors.getrow(i) # get the ith row as coordinate - value pairs
    cx = coo_matrix(individual_tweet)

    feature_tfidf = zip(cx.col, cx.data) # col is feature, data is tfidf
    # sort by tfidf, which is cx.data, in descending order and return only the top one
    top_1 = sorted(feature_tfidf, key=lambda tfidf: tfidf[1], reverse=True)[:1]

    for x in top_1:         
        top_tweet_list.append([index_vocab[x[0]],x[1]])
pprint(sorted(top_tweet_list, key=lambda tfidf: tfidf[1], reverse=True))
```

```python
# scaling not necessary for text data, since cosine distance will take care of this

# from sklearn.preprocessing import scale
# tweet_vectors = scale(tweet_vectors.todense())
```

```python
from sklearn.cluster import KMeans

cluster_model = KMeans(n_clusters = 5)
clusters = cluster_model.fit_predict(tweet_vectors)

# .fit - provides just an object 
# fit_predict - provides an array

print 'predict', clusters
```

```python
# zip clusters and the sentences/tweets
# append to list

cluster_tweet = zip(clusters, tweet_list)
```

```python
cluster_tweet_list = [tweets for tweets in cluster_tweet]
```

```python
cluster_tweet_list = sorted(cluster_tweet_list, key = lambda x: x[0])
```

```python
from pprint import pprint
pprint(cluster_tweet_list)
```

```python
# try with 6 clusters
cluster_model = KMeans(n_clusters = 6)
clusters = cluster_model.fit_predict(tweet_vectors)
cluster_tweet = zip(clusters, tweet_list)
cluster_tweet_list = [tweets for tweets in cluster_tweet]
cluster_tweet_list = sorted(cluster_tweet_list, key = lambda x: x[0])
```

### Challenge 2

Draw the inertia curve over different k values. (Sklearn KMeans class has an inertia_ attribute)

```python
K = range(1,20)

inertias = []
for k in K:
    # create a new KMeans object for each value of k
    kmeans = KMeans(n_clusters=k, random_state=8)
    kmeans.fit(tweet_vectors)
    inertias.append(kmeans.inertia_) # inertia: Sum of distances of samples to their closest cluster center
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


num_blobs = 5
ideal_k = num_blobs -1 # (we know in advance that this is the ideal value for k)

# plot elbow curve
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(K, inertias, 'b*-')

# plot the red circle
ax.plot(K[ideal_k], inertias[ideal_k], marker='o', markersize=12, 
      markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Sum of distances of samples to their closest cluster center')
tt = plt.title('Elbow for K-Means clustering') 
```
